```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(1,7,50) #creates a vector of 50 evenly spaced values between 1 and 7
y = 3 + 2*x + 1.5*np.random.randn(len(x))
df = pd.DataFrame({"xData":x,"yData":y})
sns.regplot(x="xData",y="yData",data=df) #creates plot with points, the best-fit line and the confidence interval  
```
