```python
# -*- coding: utf-8 -*- #setting the coding
import numpy as np #importing packages
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,10,0.1) #creates vector from 0 to 10 with step 0.1
freq = 0.5 #creates a variable 
y = np.sin(2*np.pi*freq*t) #as np.sin is numpy command it works for every value in vector t
```

```python
plt.plot(t,y) #creates plot of function y=sin(t)
plt.xlabel("Time[sec]") #adds x label to the plot
plt.ylabel("Sin values") #adds y label to the plot
plt.show() #shows the plot on screen (probably doesn't matter in jupyter notebook)
```

```python
def squared(x): #definition of a function in python
    return (x,x*x) #functions can return single values or other data types, e.g. tuples
y = squared(5)
print(y)
```
