```python
#univariate data means data with only one variable
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns
```

```python
# Generate the data
x = np.random.randn(500)
```

```python
#scatter plot
plt.plot(x,".") # plot the data as dots
```

```python
#scatter plot
plt.scatter(np.arange(len(x)),x,s=10) #s is responsible for size of markers
#help(plt.scatter)
plt.title("Scatterplot")
plt.xlabel("Datapoints")
plt.ylabel("Values")
```

```python
#histogram
plt.hist(x,bins=20)
plt.xlabel("Data values")
plt.ylabel("Frequency")
plt.title("Histogram")
#help(plt.hist)
```

```python
#histogram
plt.hist(x,density=True,bins=20) #if density=True we change y axis to density
plt.xlabel("Data values")
plt.ylabel("Density")
plt.title("Histogram")
#help(plt.hist)
```

```python
#histogram
sns.histplot(data=x,bins=20,color="green",element="poly") #element=" " changes looks of the plot
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
```

```python
#histogram
sns.histplot(data=x,bins=20,color="green",kde=True) #kde is responsible for the density line
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
```

```python
#density plot, based on kernel density estimation
x2 = np.random.randn(100) #change number of data points to be able to see the rugplot clearly
sns.kdeplot(data=x2,)
plt.xlabel("Values")
plt.title("Density plot")
#rugplot
sns.rugplot(data=x2,color="red") #rugplot is a plot that shows each data element as a tick mark on the x-axis
```

```python
np.random.seed(123)
#nd = stats.norm(0,1)
#normal = nd.rvs(500)
normal = np.random.normal(0,1,500)
sns.displot(normal,kind="kde",color="black") #plots "real" density of a normal distribution
sns.kdeplot(x,color="red",bw_method=0.1,linestyle="dashed") #bw is bandwidth, the smaller the more detailed the plot
sns.kdeplot(x,color="green",bw_method=1,linestyle="dotted")
sns.kdeplot(x,color="blue",bw_method=0.4)
#the bandwidth is responsible for the smoothness of the plot
#we can easilly see that the bandwidth of 0.1 is too small and the bandwidth of 1 is too big, making the plot undersmoothed or oversmoothed
#the bandwidth of 0.4 is closer to being right
```

```python
sns.displot(normal,kde=True)
```

```python
#cumulative frequency (dystrybuanta)
sns.kdeplot(x,cumulative=True)
plt.xlabel("Values")
plt.title("Cumulative frequency")
```

```python
#cumulative frequency
#different aproach that on y axis shows frequency as numbers, x axis doesn't make much sense though?
plt.plot(stats.cumfreq(x)[0])
#print(stats.cumfreq(x))
```

```python
#error-bars
#show standard error or standard deviation of the data
#it must be stated explicitly to which one our plot should corespond
#help(plt.errorbar)
x3 = x2[0:20]
x3_std = np.std(x3)
plt.errorbar(np.arange(len(x3)),x3,yerr=x3_std,fmt="o",capsize=5,capthick=3)
#error bar shows 1) values on cartesian plain 2) y or x error that must be defined as some value
#capsize is responsible for lenght of caps (lines at the end of error bars)
#capthick is reponsible for the thinckness of caps
```

```python
plt.errorbar(np.arange(len(x3)),x3,xerr=0.5,fmt="o",capsize=4,capthick=1)
```

```python
#boxplot
#that shows median, 1st and 3rd quartile
#whiskers show values in intervals:
#top whisker (Q3; Q3+1.5*IRQ>
#bottom whisker <Q1-1.5*IRQ;Q1)
#IRQ (interquartile range) = Q3 - Q1
#any value outside those intervals is called outlier and is ploted separatelly  
plt.boxplot(x3,sym=".")
```

```python
#violion plot
#violin plots show some information like boxplot, namely the box and whiskers
#additionally, it draws kde function symetrically around the box to show density of data
#np.random.seed(125)
normal1 = np.random.normal(0,1,100)
normal2 = np.random.random(100)
df_normal = pd.DataFrame({"Data1":normal1,"Data2":normal2})
sns.violinplot(df_normal)
```

```python
#grouped bar charts
#we create a data frame with 4 columns and 5 rows
df = pd.DataFrame(np.random.rand(5,4), columns=["a","b","c","d"])
print(df)
#we can plot our data and as it is based on data frame, it will be grouped by columns
df.plot(kind="bar",grid=True)
#grid=True adds a grid to the plot
```

```python
#pie chart
fractions = np.array([[30,20,15,30,5]])
df = pd.DataFrame(fractions,columns=["Dogs","Cats","Frogs","Horses","Monkeys"])
print(df.values)
plt.pie(df.values[0], explode=[0,0.1,0.15,0,0],labels=df.columns,autopct="%1.1f%%",shadow=False,startangle=180,colors=sns.color_palette("muted"))
#first argument corresponds to values of pie chart
#argument explode says how far from the center the slices should be
#labels defines labels of pie slices
#autopct defines the format of numbers
#shadow says if there should be shadow below slices or not
#startangle ???
#colors allow to change color pallete used in the plot
```
