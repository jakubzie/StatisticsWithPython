Continuous distributions can be characterized by their location and scale. For example, for normal distribution, location would be mean and scale would be standard deviation.\
\
Location:\
A location parameter $x_0$ determines the location of the shift of a distribution.
$$f_{x0}(x) = f(x-x_0)$$
Examples of location parameters could be the mean, the median or the mode.  

Scale:\
The scale parameter describes the width of the distribution. If the scale parameter is large, the distribution will be more spread out. If it's small, then the distribution will be more concentrated.\
\
Shape:\
To all other parameters beyond location and scale we refer to as shape parameters. Almost all of distributions we use in statistics have only one or two parameters.\
\
Skeweness:\
Skeweness is a statistic that can answer the question of the distribution being symmetrical.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import scipy.stats as stats
warnings.filterwarnings("ignore")
```

Kurtosis:\
Kurtosis is a measure of the "peakedness" of the probability distribution. Normals distribution has a kurtosis of 3. We normally calculate something called excess kurtosis, which is equal to kurtosis - 3. Excess kurtosis is equal to 0 for normal distribution. If excess kurtosis is bigger than 0 the distribution is called platykurtic, if it's smaller the distribution is called leptokurtic. 

In Python the most elegant way to work with distributions is a two step procedure:
1. Define distribution (a frozen distribution)
2. Calculate desired statistics based on the frozen distribution.

```python
nd = stats.norm(0,1)
x = np.linspace(-4,4,100)
pdf = nd.pdf(x)
cdf = nd.cdf(x)
plt.plot(pdf)
plt.plot(cdf)
```
